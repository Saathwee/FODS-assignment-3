```python
#1. abs()-Returns the absolute value of a number
x = abs(-7.25)
print("Abs: ",x)

#2. all()-Returns True if all items in an iterable object are true
mylist = [True, True, True]
x = all(mylist)
print("All: ",x)

#3. bool()-Returns the boolean value of the specified object
x = bool(1)
print("Bool: ",x)

#4. chr()-Returns a character from the specified Unicode code.
x = chr(100)
print("Chr: ",x)

#5. complex()-Returns a complex number
x = complex(3, 5)
print("Complex: ",x)

#6. dict()-Returns a dictionary (Array)
x = dict(name = "Joe", age = 13, country = "India")
print("Dict: ",x)

#7. eval()-Evaluates and executes an expression
x = 'print(55)'
print("Eval: ",x)

#8. float()-Returns a floating point number
x = float(3)
print("Float: ",x)

#9. hex()-Converts a number into a hexadecimal value
x = hex(255)
print("Hex: ",x)

#10. id()-Returns the id of an object
x = ('apple', 'banana', 'cherry')
y = id(x)
print("ID: ",y)

#11. input()-Allowing user input
x = int(input("Number: "))
print("input: ",x)

#12. int()-Returns an integer number
x = int(3.5)
print("int: ",x)

#13. len()-Returns the length of an object
mylist = ["apple", "orange", "cherry"]
x = len(mylist)
print("len: ", x)

#14. list()-Returns a list
x = list(('apple', 'banana', 'cherry'))
print("list: ",x)

#15. max()-Returns the largest item in an iterable
x = max(5, 10)
print("Max :",x)

#16. min()-Returns the smallest item in an iterable
x = min(5, 10)
print("Min: ",x)

#17. pow()-Returns the value of x to the power of y
x = pow(4, 3)
print("Power: ",x)

#18. round()-Rounds a numbers
x = round(5.76543, 2)
print("Round: ",x)

#19. sorted()-Returns a sorted list
a = ("b", "g", "a", "d", "f", "c", "h", "e")
x = sorted(a)
print("Sorted: ",x)

#20. sum()-Sums the items of an iterator
a = (1, 2, 3, 4, 5)
x = sum(a)
print("Sum: ",x)
```
